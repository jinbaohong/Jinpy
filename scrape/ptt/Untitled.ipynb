```python
def getMainContent(mainDiv):
    contentList = []
    tmpContent = ''
    for tag in mainDiv:
        if (type(tag) is bs4.element.NavigableString):
            tmpContent += tag
        elif tag.has_attr('class'):
            if tag['class'][0] == 'f6':
                tmpContent += tag.text
            elif tag['class'][0] == 'f2':
                tmpContent += tag.text
            elif tag['class'][0] == 'push':
                if tmpContent: contentList.append({'type': 'main',
                                                   'content': tmpContent})
                tmpContent = ''
                contentList.append({'type': 'push',
                                    'content': {
                                        'pushTag': tag.findAll('span')[0].text,
                                        'pushUserid': tag.findAll('span')[1].text,
                                        'pushContent': tag.findAll('span')[2].text,
                                        'pushIpDateTime': tag.findAll('span')[3].text}
                                   })
        elif tag.has_attr('href'):
            tmpContent += tag['href']
    if tmpContent: contentList.append({'type': 'main',
                                       'content': tmpContent})
    return contentList

def setDict(soup, url):
    result = dict()
    mainDiv = soup.find('div', {'id':'main-content'})
    metaInfos = mainDiv.findAll('div', {'class':'article-metaline'})
    try:
        result['author'] = metaInfos[0].findAll('span')[1].text
    except:
        result['author'] = 'Unknown'
    try:
        result['title'] = metaInfos[1].findAll('span')[1].text
    except:
        result['title'] = 'Unknown'
    try:
        result['timeStamp'] = metaInfos[2].findAll('span')[1].text
    except:
        result['timeStamp'] = 'Unknown'
    try:
        result['mainContent'] = getMainContent(mainDiv)
    except:
        result['mainContent'] = []
    result['length'] = len(result['mainContent'])
    result['uri'] = url
    return result

url = 'https://www.ptt.cc/bbs/intltrade/M.1520819570.A.03E.html'
url = 'https://www.ptt.cc/bbs/intltrade/M.1520146741.A.497.html'
url = 'https://www.ptt.cc/bbs/intltrade/M.1136385844.A.A43.html'
url = 'https://www.ptt.cc/bbs/intltrade/M.1499253026.A.338.html'
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')
setDict(soup, url)
```
